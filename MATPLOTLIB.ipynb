```python
# how to make two plots together
plt.figure# initialize a page
plt.subplot(1,2,1) # insight a page having the sub part means 1 row 2 columns and last 1 is for acceess the first
plt.plot(x, np.sin(x))

plt.subplot(1, 2, 2)# this ending 2 is for the access the second column
plt.plot(x, np.cos(x))

plt.show()
```

```python
# this is for the two rows and 1 column
# first last 1 is for access the first 
# and the second paragrapg last 2 is for access the second row
plt.figure() #initialising a page to make plot
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
# this is for the the 2 rows and 2 column means 4 graphs in one page
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))

plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x))

plt.subplot(2, 2, 4)
plt.plot(x, np.sin(x+1))

plt.show()
```

```python
#pie char >> To understand the percent distribution of data
langs = ["C", "Python", "Java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%")
plt.title("Pie chart")
plt.show()
```

```python
#focus on one segment
explode = (0.0, 0.1, 0.1)
langs = ["C", "Python", "Java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%", explode = explode)
plt.title("Pie chart")
plt.show()
```

```python
explode = (0.0, 0.1, 0.1)
langs = ["C", "Python", "Java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%", explode = explode, shadow = True)
plt.title("Pie chart")
plt.show()
```

```python
years = [2021, 2018, 2022, 2023, 2017]
category1 = [20, 25, 30, 28, 35]
category2 = [10, 15, 20, 18, 25]
category3 = [5, 8, 10, 9, 15]

plt.stackplot(years, category1, category2, category3, labels = ["category1", "category2", "category3",],
             colors = ["skyblue","lightgreen", "orange"])
plt.xlabel("years")
plt.ylabel("Values")
plt.title("Area plot")
plt.legend()
plt.show()
```
