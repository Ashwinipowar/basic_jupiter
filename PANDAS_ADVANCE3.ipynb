```python
#merge
df1= pd.DataFrame({'key': [1,2,4,5,6],
                   'key2':[4,5,6,7,8],
                   'key3':[3,4,5,6,6]
                  }) 
df1
```

```python
df2 =pd.DataFrame({ 'key':[1,2,45,6,67],
                   'key4':[56,5,6,7,8],
                   'key5':[3,56,5,6,6]})
df2
```

```python
pd.merge(df1,df2,how='inner')
```

```python
pd.merge(df1,df2,how = 'left')
```

```python
pd.merge( df1,  df2, how='right')
```

```python
# all the common element of the common column from both the dataframe
pd.merge(df1, df2, how='outer')
```

```python
#merge
df1= pd.DataFrame({ 'key1': [1,2,4,5,6],
                   'key2':[4,5,6,7,8],
                   'key3':[3,4,5,6,6]},
                    index =['a','b','c','d','e']) 
df1
```

```python
pd.merge(df1,df2, how="cross")
```
