```python
# ðŸ”¹ What is shutil in Python?
# shutil stands for shell utilities.
# It's a built-in Python module that helps you perform file operations like:

# copying files and directories

# moving or renaming files

# deleting files or folders

# creating archives (zip, tar, etc.)

# ðŸ”¹ Commonly Used shutil Functions
# Function	Description
# shutil.copy(src, dst)	Copies a file from src to dst
# shutil.copytree(src, dst)	Copies an entire folder (recursively)
# shutil.move(src, dst)	Moves a file or directory
# shutil.rmtree(path)	Deletes an entire folder tree
# shutil.make_archive()	Creates zip/tar archives
```
